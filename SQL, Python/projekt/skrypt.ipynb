```python
import pandas as pd
import numpy as np
```

```python
storey_species = pd.read_csv("f_storey_species.txt", sep = "\t")
```

```python
species = pd.read_csv("f_tree_species_dic.txt", sep = "\t")
```

```python
storey_species = pd.merge(storey_species, species[['species_cd','species_name']], on = 'species_cd', how = 'left')
```

```python
storey_species
```

```python
df = storey_species[~storey_species['volume'].isna()]
```

```python
df
```

```python
species_groups = df.groupby('species_cd')
```

```python
species_groups[['species_name']].first()
```

```python
df = df.pivot_table(index = 'arodes_int_num', columns = 'species_name', values = 'volume', aggfunc = 'sum')
```

```python
df
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
topola = df['topola biała'].fillna(0) + df['topola osika'].fillna(0)
wierzba = df['wierzba biała'].fillna(0) + df['wierzba iwa'].fillna(0)
```

```python
df = df.drop(columns = ['topola biała', 'topola osika', 'wierzba biała', 'wierzba iwa'])
```

```python
df['topola'] = topola
df['wierzba'] = wierzba
```

```python
df
```

```python
df = df.replace(0, np.nan)
```

```python
df2 = df.notnull()
```

```python
df2
```

```python
subarea = pd.read_csv("f_subarea.txt", sep = "\t")
```

```python
subarea
```

```python
subarea.columns = subarea.columns.str.strip()
```

```python
df2 = pd.merge(df2, subarea[['arodes_int_num', 'soil_subtype_cd', 'forest_func_cd', 'moisture_cd', 'site_type_cd']], on = 'arodes_int_num', how = 'left')
```

```python
df2
```

```python
df2[['site_type_cd']] = df2[['site_type_cd']].replace(to_replace = r'\bB\w+', value = 'bór', regex = True)
df2[['site_type_cd']] = df2[['site_type_cd']].replace(to_replace = r'\bL\w+', value = 'las', regex = True)
df2[['site_type_cd']] = df2[['site_type_cd']].replace(to_replace = r'\bO\w+', value = 'ols', regex = True)
```

```python
soil_groups = df2.groupby('soil_subtype_cd').size().to_frame(name = 'count').reset_index()
```

```python
soil_groups
```

```python
soils = pd.read_csv("f_soil_subtype_dic.txt", sep = "\t")
soils
```

```python
soil_groups = pd.merge(soil_groups, soils[['soil_subtype_cd', 'soil_subtype_name']], on = 'soil_subtype_cd', how = 'left')
```

```python
soil_groups
```

```python
df2['soil_subtype_cd'] = df2['soil_subtype_cd'].str.strip()
df2['forest_func_cd'] = df2['forest_func_cd'].str.strip()
df2['moisture_cd'] = df2['moisture_cd'].str.strip()
df2['site_type_cd'] = df2['site_type_cd'].str.strip()
```

```python
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('B', 'bielicowe')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('Bw', 'bielicowe')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('BR', 'brunatne')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bBR\w+', value = 'brunatne', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bBg\w+', value = 'glejo_bielicowe', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bCZ\w+', value = 'czarne_ziemie', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bD\w+', value = 'deluwialne', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bG\w+', value = 'gruntowoglejowe', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('MRm', 'mineralno_murszowe')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('MRms', 'murszaste')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('MRw', 'murszowate')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace('Mt', 'torfowo_murszowe')
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bOG\w+', value = 'opadowoglejowe', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bP\w+', value = 'płowe', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bRD\w+', value = 'rdzawe', regex = True)
df2[['soil_subtype_cd']] = df2[['soil_subtype_cd']].replace(to_replace = r'\bT\w+', value = 'torfowe', regex = True)
```

```python
df2
```

```python
df2.rename(columns = {'soil_subtype_cd':'gleba', 'forest_func_cd':'funkcja_lasu', 
                      'moisture_cd':'uwilgotnienie', 'site_type_cd':'siedlisko' }, inplace = True)
```
