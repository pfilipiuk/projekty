```python
df1['Name'] = df1['Name'].str.strip()
df1['Address'] = df1['Address'].str.strip()
df1['city'] = df1['city'].str.strip()
df1['state'] = df1['state'].str.strip()
df1['ContactPref'] = df1['ContactPref'].str.strip()
df1['Contact'] = df1['Contact'].str.strip()
```

```python
df2['cusnum'] = df2['cusnum'].str.strip()
df2['cusfname'] = df2['cusfname'].str.strip()
df2['cuslname'] = df2['cuslname'].str.strip()
df2['cusmiddle'] = df2['cusmiddle'].str.strip()
df2['cusemail'] = df2['cusemail'].str.strip()
df2['cusphone'] = df2['cusphone'].str.strip()
df2['cusaddress1'] = df2['cusaddress1'].str.strip()
df2['cusaddress2'] = df2['cusaddress2'].str.strip()
df2['cuscity'] = df2['cuscity'].str.strip()
df2['cusstate'] = df2['cusstate'].str.strip()
df2['cuszip'] = df2['cuszip'].str.strip()
```
